```python

average_df=Site_Loads_new.groupby(['Site_Id'])['Units'].mean()
average_df = pd.merge(left=Site_Loads_new, right=average_df, on='Site_Id', how='left')
average_df = average_df.rename(columns={'Units_y': 'avg_Units','Units_x': 'Units'})
average_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

average_df['error'] = average_df['Units'] - average_df['avg_Units']
average_df['error_percentage'] = average_df['error'] / average_df['Units'] * 100

average_df['status'] = 'dropped'

average_df.loc[(average_df['error_percentage'] > -80) & (average_df['error_percentage'] < 80), 'status'] = 'taken'
```

### remove power shareable sites

```python

ids_to_remove = ['AM0008','AM0035','AM0088','AM0122','AM0142','AN0002','AN0013','AN0015','AN0016','AN0095','AN0222','BA0013','BA0023','BA0029','BA0042','BA0066','BA0084','BA0114','BD0001','BD0010','BD0029','BD0044','BD0120','CM0009','CM0053','CM0054','CM0056','CM0121','CM0175','CM0220','CM0238','CM0249','CM0256','CM0260','CM0312','CM0339','CM0351','CM0386','CM0417','CM0450','CM0463','CM0502','CM0590','CM0623','CM0667','CM0742','CM0823','CM0827','CM0967','CM1029','CM1034','CM1035','CM1077','CM1168','CM1222','CM1226','CM1338','CM1463','GA0006','GA0020','GA0030','GA0041','GA0054','GA0097','GA0105','GA0110','GA0182','GM0052','GM0084','GM0094']

# Select only the rows that do not have any of the IDs you want to remove
site_info = site_info[~site_info['Site_Id'].isin(ids_to_remove)]

site_info['day_of_week'] = pd.to_datetime(site_info['Time_Key'], format='%Y%m%d%H').dt.dayofweek
site_info['hour'] = pd.to_datetime(site_info['Time_Key'], format='%Y%m%d%H').dt.hour
```

```python
site_info = site_info.drop(columns=['status','error','error_percentage'])
# df_merged_final = df_merged_final.rename(columns={'Units_x': 'Units'})
```

```python
zerofill_columns = ['Medium_Pwr_RRU_Count','Low_Pwr_RRU_Count','High_Pwr_RRU_Count', 'Router_Count', 'Links_Count']

for col in zerofill_columns:
    site_info[col] = site_info[col].fillna(0)

site_info.count()
```

```python
site_info = site_info[['Site_Id','Time_Key','Units','hour', 'Router_Count', 'Links_Count', 'High_Pwr_RRU_Count','Medium_Pwr_RRU_Count','Low_Pwr_RRU_Count','Cabin_Type','Site_Type','District','Region','day_of_week']]
site_info

def cabintype_correction(inpt):
    if inpt == 'Lamp pole' or inpt == 'Lamp Pole' or inpt == 'lamp Pole':
        return 'Lamp Pole'
    else:
        return inpt
```

```python
site_info['Cabin_Type'] = site_info['Cabin_Type'].apply(cabintype_correction)
site_info["Cabin_Type"] = site_info["Cabin_Type"].replace("\s+$", "", regex=True)
site_info = site_info.dropna()
site_info = site_info[site_info["Region"].apply(lambda x: not isinstance(x, int))]
```
