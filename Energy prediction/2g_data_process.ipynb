```python
import pandas as pd
huawei=pd.read_csv('../data/raw/latest-2022-12/2G Traffic Volume/huawei2g.csv',skiprows=range(0,6))
# ericssion=pd.read_excel('../data/raw/latest-2022-12/2G Traffic Volume/ericssion2g.xlsx')
```

```python
# read in the Excel files

ericssion1=  pd.read_csv('../data/raw/latest-2022-12/2G Traffic Volume/ericssion2g1.csv')
ericssion2 = pd.read_csv('../data/raw/latest-2022-12/2G Traffic Volume/ericssion2g2.csv')
ericssion3 = pd.read_csv('../data/raw/latest-2022-12/2G Traffic Volume/ericssion2g3.csv')
ericssion4 = pd.read_csv('../data/raw/latest-2022-12/2G Traffic Volume/ericssion2g4.csv')

# concatenate the dataframes

ericssion = pd.concat([ericssion1, ericssion2, ericssion3, ericssion4])
```

```python
cell_inventory=pd.read_csv('../data/raw/latest-2022-12/2G Traffic Volume/Cell Inventory.csv')
```

```python
huawei=huawei[['Time','Site Name','L9503:Total Bytes of Downlink LLC PDUs Received(byte)','L9403:Total Bytes of Uplink LLC PDUs Sent(byte)']]
huawei=huawei.rename(columns={'Site Name':'Cell_Name','L9503:Total Bytes of Downlink LLC PDUs Received(byte)':'dl2gtraffic','L9403:Total Bytes of Uplink LLC PDUs Sent(byte)':'ul2gtraffic'})
huawei['Site_Id']=huawei['Cell_Name'].str.split("-").str[1]
huawei=huawei.drop(['Cell_Name'],axis=1)

huawei=huawei.rename(columns={'Time':'Date'})
huawei['Date'] = pd.to_datetime(huawei['Date'], format='%Y-%m-%d %H:%M', errors='coerce')
huawei = huawei.dropna()

huawei['Date'] = pd.to_datetime(huawei['Date'], format='%Y-%m-%d %H:%M')
huawei['Time_Key'] = huawei['Date'].apply(lambda x: int(x.strftime('%Y%m%d%H')))
huawei['dl2gtraffic']=huawei['dl2gtraffic']/1000000
huawei['ul2gtraffic']=huawei['ul2gtraffic']/1000000
huawei = huawei.reindex(columns=['Site_Id', 'Time_Key', 'dl2gtraffic', 'ul2gtraffic'])
```

```python
huawei
```

```python
huawei = huawei.groupby(['Site_Id','Time_Key']).sum()
huawei=huawei.reset_index()
huawei
```
