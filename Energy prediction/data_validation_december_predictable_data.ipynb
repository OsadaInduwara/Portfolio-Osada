```python
merge = pd.merge(traffic,all_sites_to_be_predicted,on='Site_Id')
```

```python
# merge['Date'] = pd.to_datetime(merge['Date'])
# merge['day_of_week'] = merge['Date'].dt.dayofweek


merge['DT'] = pd.to_datetime(merge['Time_Key'], format='%Y%m%d%H')
merge['day_of_week'] = merge['DT'].dt.strftime('%w')
merge['hour'] = merge['DT'].dt.strftime('%H')
```

```python
merge
```

```python
# print(december_predictable_data.isna().any())
def cabintype_correction(inpt):
    if inpt == 'Lamp pole' or inpt == 'Lamp Pole' or inpt == 'lamp Pole':
        return 'Lamp Pole'
    else:
        return inpt
```

```python
zerofill_columns = ['Medium_Pwr_RRU_Count','Low_Pwr_RRU_Count','High_Pwr_RRU_Count', 'Router_Count', 'Links_Count']

for col in zerofill_columns:
    merge[col] = merge[col].fillna(0)

merge.count()
```

```python
merge['Cabin_Type'] = merge['Cabin_Type'].apply(cabintype_correction)
merge["Cabin_Type"] = merge["Cabin_Type"].replace("\s+$", "", regex=True)
merge = merge.dropna()
merge = merge[merge["Region"].apply(lambda x: not isinstance(x, int))]
```

```python
merge = merge[merge.Cabin_Type != 'Lamp Pole']
```

```python
merge['District'] = merge['Site_Id'].str[:2]
```

```python
ids_to_remove = ['AM0008','AM0035','AM0088','AM0122','AM0142','AN0002','AN0013','AN0015','AN0016','AN0095','AN0222','BA0013','BA0023','BA0029','BA0042','BA0066','BA0084','BA0114','BD0001','BD0010','BD0029','BD0044','BD0120','CM0009','CM0053','CM0054','CM0056','CM0121','CM0175','CM0220','CM0238','CM0249','CM0256','CM0260','CM0312','CM0339','CM0351','CM0386','CM0417','CM0450','CM0463','CM0502','CM0590','CM0623','CM0667','CM0742','CM0823','CM0827','CM0967','CM1029','CM1034','CM1035','CM1077','CM1168','CM1222','CM1226','CM1338','CM1463','GA0006','GA0020','GA0030','GA0041','GA0054','GA0097','GA0105','GA0110','GA0182','GM0052','GM0084','GM0094']

# Select only the rows that do not have any of the IDs you want to remove
merge = merge[~merge['Site_Id'].isin(ids_to_remove)]
```
