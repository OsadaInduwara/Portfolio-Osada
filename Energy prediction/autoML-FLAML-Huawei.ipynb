```python
import pandas as pd
import numpy as np
```

```python
training_site_info = pd.read_csv('../data/processed/final_dataset6.csv')
```

```python
training_site_info=training_site_info.drop(['Unnamed: 0'],axis=1)
```

```python
training_site_info=training_site_info[training_site_info['Site_Type'] == 'Huawei']
training_site_info
```

```python
# training_site_info['Site_Id'].count()
training_site_info=training_site_info.drop('Site_Type',axis=1)
training_site_info
```

```python
# Set a random seed to ensure reproducibility
random_state = 42

# Randomly select 20% of the rows
sample = training_site_info.sample(frac=0.2, random_state=random_state)

# Separate the sampled rows into a new DataFrame
sampled_data = training_site_info.loc[sample.index]
training_site_info1 = training_site_info.drop(sample.index)
```

```python
training_site_info1
```

```python
sampled_data
```

```python
excluded_columns = ['Site_Id', 'Time_Key']
training_site_info_dummies1 = training_site_info1.loc[:, training_site_info1.columns.difference(excluded_columns)].pipe(pd.get_dummies)

excluded_columns = ['Site_Id', 'Time_Key']
training_site_info_dummies = training_site_info.loc[:, training_site_info.columns.difference(excluded_columns)].pipe(pd.get_dummies)

excluded_columns = ['Site_Id', 'Time_Key']
sampled_data_dumies = sampled_data.loc[:, sampled_data.columns.difference(excluded_columns)].pipe(pd.get_dummies)
```

```python
from flaml import AutoML
from sklearn.model_selection import train_test_split

X = training_site_info_dummies.drop('Units', axis=1)
y = training_site_info_dummies['Units']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Initialize the AutoML object
automl = AutoML()

# Fit the model to the training data
automl.fit(X_train, y_train,task="regression", time_budget=1000)

# Evaluate the model on the test data
test_score = automl.score(X_test, y_test)
print("Test score: ", test_score)
```
